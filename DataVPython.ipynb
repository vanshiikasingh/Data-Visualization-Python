```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
df = pd.read_json("rain.json")
print(df)
```

```python
print("df statistics: ", df.describe())
```

```python
df.plot(x= 'Month', y='Temperature')
df.plot(x='Month', y='Rainfall')
```

MatPlotLib Plot

```python
#create the dataframe using the json file
plt.figure(figsize=(15,5))
plt.plot(df['Month'], df['Temperature'], label='Temperature')
plt.show()
```

```python
#plot rainfall
plt.figure(figsize=(15,5))
plt.plot(df['Month'], df['Temperature'], label='Rainfall')
plt.show()
```

```python
#ploting rainfall and temp together
plt.figure(figsize=(15,5))
plt.plot(df['Month'], df['Rainfall'], label='Rainfall')
plt.plot(df['Month'], df['Temperature'], label='Temperature')
plt.legend() #shows diff color
plt.show()
```

Scatterplot Graph (Seaborn)

```python
df = pd.read_csv("tempYearly.csv")
print(df)
```

```python
sns.set(rc={'figure.figsize':(12,6)}) #to change size
sns.scatterplot(x='Year', y= 'Temperature', data= df)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,6)}) #to change size
sns.regplot(x='Year', y= 'Temperature', data= df)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,6)})
sns.regplot(x='Rainfall', y= 'Temperature', data= df)
plt.show()   
```

HeapMap Plot

```python
data = pd.read_csv("birthYearly.csv")
print(data)
```

```python
dataP = data.pivot("month", "year", "births")
print(dataP)
sns.heatmap(dataP, annot= True, fmt="d")
plt.show()
```

Jointplot Distribution

```python
data = pd.read_csv("tempYearly.csv")
print(data)
```

```python
#histogram tells where the data is intense
sns.jointplot("Rainfall", "Temperature", data= data)
plt.show()
```

```python
sns.jointplot("Rainfall", "Temperature", data= data, kind = "hex")
plt.show()
```

```python
sns.jointplot("Rainfall", "Temperature", data= data, kind= "reg")
plt.show()
```
